# Currency Converter using Python and Fixer API

This project fetches real-time exchange rates using the Fixer API and converts one currency to another.

## Features
- Real-time currency conversion
- Supports multiple currencies
- User input based conversion
- Error handling for invalid currency codes

# Import Libraries

```python
import requests #fetch data from API
import json #parse API response
import sys #exit program
from pprint import pprint #neatly print currency list
```

```python
url = "http://data.fixer.io/api/latest?access_key="

data = requests.get(url).text
data2 = json.loads(data)

fx = data2["rates"]

print("Exchange rates loaded successfully!")
```

What this does:
- onnects to Fixer API
- Gets latest currency rates
- Stores rates in fx


Example : fx["INR"]

which gives Indian Rupee rate.

# Currency Converter Function

```python
def currency_converter():

    query = input(
        "Enter amount, from currency, to currency\nExample: 100 USD INR\n"
        "Type SHOW for currencies\nType Q to quit\n"
    )

    if query.upper() == "Q":
        return

    elif query.upper() == "SHOW":
        pprint(currencies)
        currency_converter()

    else:
        try:
            qty, fromC, toC = query.split()

            qty = float(qty)
            fromC = fromC.upper()
            toC = toC.upper()

            amount = round(
                qty * fx[toC] / fx[fromC],
                2
            )

            print(
                f"{qty} {fromC} = {amount} {toC}"
            )

        except KeyError:
            print("Invalid currency code.")

        except ValueError:
            print("Wrong input format.")
```
